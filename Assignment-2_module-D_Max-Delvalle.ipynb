The `desc` field contains text descriptions of loans. Tokenize each loan description and count the words for 'good' and 'bad' loan descriptions, putting counts into two separate `Counter()` data structures according to each loan's status in `statuses`. Print out the 50 most common words used to describe 'good' and 'bad' loans, respectively. 

__Note__: If you're still working on __D1__, for partial (near-full) credit you can still count all words, regardless of loan good/bad status, and print out the 50 most common.

```python
Desc_Loans = {"Good": [], "Bad": []}
for j, row in enumerate(loan_lists):
    if j:
        desc = row[19]
        status = row[16]
        if status=='Current'or status=='Fully Paid'or status=='Issued':
            Desc_Loans["Good"].append(row[19])
        else:
            Desc_Loans["Bad"].append(row[19])
    else:
        header = row 
        
totalgoodstring = ""
totalbadstring = ""

#Below we iterate the split for each Good and Bad loans:

for eachentry in Desc_Loans['Good']:
    if len(eachentry.split(" > "))==2:
        totalgoodstring += eachentry.split(" > ")[1]
    else:
        totalgoodstring += ""
goodstring = re.sub("<br>", " ", totalgoodstring.lower())

for eachentry in Desc_Loans['Bad']:
    if len(eachentry.split(" > "))==2:
        totalbadstring += eachentry.split(" > ")[1]
    else:
        totalbadstring += ""
badstring = re.sub("<br>", " ", totalbadstring.lower())

print("Good String is:\n "+goodstring[:500]+" ...")
print("Bad String is:\n "+badstring[:500]+" ...")
```

```python
GoodWords = goodstring.split()
BadWords = badstring.split()

GoodWordCounter = Counter(GoodWords)
for w in sorted(GoodWordCounter, key=GoodWordCounter.get, reverse=True)[:50]:
    print((w, GoodWordCounter[w]))
```

```python
BadWordCounter = Counter(BadWords)
for w in sorted(BadWordCounter, key=BadWordCounter.get, reverse=True)[:50]:
    print((w, BadWordCounter[w]))
```
