__B3.__ _(6 points)_ Now, using all matches in the accessed data, construct aggregate result data for each team and store this in a list. Your final output should look like the table stored in `data/sample-table.txt`:

Here, the fields are: abbreviation of team name, games played, games won, games drawn, games lost, goals for, goals against, goal difference, and total points. 

````python
uniqueTeams = list(set(Teams)) 

GoalFor = []
for row in Game_table:
    GoalFor.append([row[0], row[1]])
    GoalFor.append([row[2], row[3]])
GoalAgainst = []
for row in Game_table:
    GoalAgainst.append([row[0], row[3]])
    GoalAgainst.append([row[2], row[1]])
````

````python
lose = []
draw = []
win = []

#For this portion, we have to decide on winners and losers

for each in Game_table:
    if each[4][0]>each[4][1]:
        win.append(each[0])
        lose.append(each[2])
    elif each[4][0]==each[4][1]:
        draw.append(each[0])
        draw.append(each[2])
    else:
        lose.append(each[0])
        win.append(each[2]) 
for each in Game_table:
    if each[5][0]>each[5][1]:
        win.append(each[0])
        lose.append(each[2])
    elif each[5][0]==each[5][1]:
        draw.append(each[0])
        draw.append(each[2])
    else:
        lose.append(each[0])
        win.append(each[2])

loserTeams = list(set(lose))
drawTeams = list(set(draw))
winnerTeams = list(set(win))
````

````python
team_table = {}
#GoalAgainst = []
for row in Game_table:
    Team = row[0]
    team_table.setdefault(Team, {})
    W = len([i for i in win if i == Team])
    D = len([i for i in draw if i == Team])
    L = len([i for i in lose if i == Team])
    G = W+D+L
    GF = sum([i[1] for i in GoalFor if i[0] == Team])
    GA = sum([i[1] for i in GoalAgainst if i[0] == Team])
    GD = GF-GA
    P = W*3+D*1
    team_table[Team]= [G, W, D, L, GF, GA, GD, P]
Soccer_List1 = [['Team', 'G', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'P']]
Soccer_List = []
for key in team_table.keys():
    info = team_table[key] 
    Soccer_List.append(
        [key, info[0], info[1], info[2], info[3], info[4], info[5], info[6], info[7]])
pprint(Soccer_List)
````
